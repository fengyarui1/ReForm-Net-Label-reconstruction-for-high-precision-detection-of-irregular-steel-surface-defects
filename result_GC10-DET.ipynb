```python
from ultralytics import YOLO
model = YOLO('runs/detect/baseline/weights/best.pt')
# 评估模型
metrics = model.val(
    data='yolo_dataset/dataset.yaml',
    imgsz=1280
)

# 打印评估指标
print(f"mAP50: {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")
```
